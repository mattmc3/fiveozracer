```python
# imports
import Foz
from Foz import read_racers, read_timer_results
from collections import namedtuple, OrderedDict
from pprint import pprint
from collections import defaultdict
from asq.initiators import query
```

```python
all_racers = read_racers()
all_racers
```

```python
all_timer_results = read_timer_results()
all_timer_results
```

```python
engine = Foz.DerbyEngine(all_racers, 7, 6)
for tr in all_timer_results:
    engine.add_timer_result(tr)
```

```python
engine.race_lineup
```

```python
tally = defaultdict(int)
for item in engine.race_lineup:
    if item.place > 0:
        tally[item.racer.car_num] += item.time

qry = query(all_racers) \
    .order_by(lambda x: x.racing_group.key) \
    .then_by(lambda x: tally[item.racer.car_num]) \
    .then_by(lambda x: x.grade.grade_num) \
    .then_by(lambda x: x.car_num) \
    .to_list()
print(qry)
```
